```python
# List of columns to be converted
columns_to_convert = [
    'SDG Index Score', '1no_poverty', '2no_hunger', '3health_wellbeing', '4quality_education',
    '5gender_equality', '6cleanwater_sanitation', '7affordable_clean_energy', '8work_economy_growth',
    '9industry_innov_infrast', '10reduced_inequalities', '11cities_communities', '12consumption_production',
    '13climate', '14life_below_water', '15life_on_land', '16peace_justice_stroninstitutions', '17partnership_for_goals'
]

# Replace commas with dots and convert to float for each column
for col in columns_to_convert:
    glob_sust_score[col] = glob_sust_score[col].str.replace(',', '.').astype('float64')
```

```python
glob_sust_score.columns
```

```python
glob_sust_score = glob_sust_score.loc[glob_sust_score['year']<=2020]
```

```python
glob_sust_score.to_csv('./sustainability/glob_sust_score.csv')
```
