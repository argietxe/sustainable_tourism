```python
employment = pd.read_csv('./untourism/employment.csv', sep=';')
```

```python
employment
```

```python
employment['Facility'].unique()
```

```python
employment.columns = ['country', 'Facility', 'Units', '2004', '2005', '2006', '2007', '2008',
       '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017',
       '2018', '2019', '2020']
```

```python
accomodations_service_hotel = employment[employment['Facility'] == 'Accommodation services for visitors (hotels and similar establishments)']
accomodation_service_other = employment[employment['Facility'] == 'Other accommodation services']
food_beverage_serving = employment[employment['Facility'] == 'Food and beverage serving activities']
passenger_transport = employment[employment['Facility'] == 'Passenger transportation']
travel_agency_and_other_resa = employment[employment['Facility'] == 'Travel agencies and other reservation services activities']
other_tourism_industry = employment[employment['Facility'] == 'Other tourism industries']
```

```python
accomodations_service_hotel = accomodations_service_hotel.drop(columns=['Facility','Units'])
accomodation_service_other = accomodation_service_other.drop(columns=['Facility','Units'])
food_beverage_serving = food_beverage_serving.drop(columns=['Facility','Units'])
passenger_transport = passenger_transport.drop(columns=['Facility','Units'])
travel_agency_and_other_resa = travel_agency_and_other_resa.drop(columns=['Facility','Units'])
other_tourism_industry = other_tourism_industry.drop(columns=['Facility','Units'])
```

```python
accomodations_service_hotel = pd.melt(accomodations_service_hotel, id_vars=['country'], var_name='year', value_name='accomodations_service_hotel')
accomodation_service_other = pd.melt(accomodation_service_other, id_vars=['country'], var_name='year', value_name='accomodation_service_other')
food_beverage_serving = pd.melt(food_beverage_serving, id_vars=['country'], var_name='year', value_name='food_beverage_serving')
passenger_transport = pd.melt(passenger_transport, id_vars=['country'], var_name='year', value_name='passenger_transport')
travel_agency_and_other_resa = pd.melt(travel_agency_and_other_resa, id_vars=['country'], var_name='year', value_name='travel_agency_and_other_resa')
other_tourism_industry = pd.melt(other_tourism_industry, id_vars=['country'], var_name='year', value_name='other_tourism_industry')
```

```python
merge1a = pd.merge(accomodations_service_hotel, accomodation_service_other, on = ['country','year'], how='right')
merge2a = pd.merge(merge1a, food_beverage_serving, on = ['country','year'], how='right')
merge3a = pd.merge(merge2a, passenger_transport, on = ['country','year'], how='right')
merge4a = pd.merge(merge3a, travel_agency_and_other_resa, on = ['country','year'], how='right')
employment = pd.merge(merge4a, other_tourism_industry, on = ['country','year'], how='right')
employment = employment.sort_values(by=['country','year'], ascending=True)
```

```python
employment
```

```python
employment.replace('..', np.nan, inplace=True)
```
