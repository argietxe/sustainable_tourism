```python
import pandas as pd
import numpy as np 
```

```python
inbound_arrivals = pd.read_csv('unwto-inbound-arrivals-data.csv', sep=';')
```

```python
inbound_arrivals
```

```python
mask1 = inbound_arrivals['del'] != 'Arrivals'
mask2 = inbound_arrivals['del1'] != 'Overnights visitors (tourists)'
mask3 = inbound_arrivals['del1'] != 'Same-day visitors (excursionists)'
mask4 = inbound_arrivals['del2'] != 'of which, cruise passengers'
combined_mask = mask1 & mask2 & mask3 & mask4
```

```python
filtered_inbound_arrivals = inbound_arrivals[combined_mask]
```

```python
filtered_inbound_arrivals
```

```python
filtered_inbound_arrivals.to_csv('filtered_inbound_arrivals.csv')
```

```python
inbound_arrivals1 = pd.read_csv('filtered_inbound_arrivals1.csv', sep=',')
```

```python
inbound_arrivals1.replace('..', np.nan, inplace=True)
```

```python
inbound_arrivals1.info()
```

```python
columns_to_convert = ['2004', '2005', '2006', '2007', '2008', '2009',
                      '2010', '2011', '2012', '2013', '2014', '2015',
                      '2016', '2017', '2018', '2019', '2020', '2021', '2022']
```

```python
for col in columns_to_convert:
    inbound_arrivals1[col] = inbound_arrivals1[col].str.replace(' ', '').str.replace(',', '.')
```

```python
for col in columns_to_convert:
    inbound_arrivals1[columns_to_convert] = inbound_arrivals1[columns_to_convert].astype(float)
```
