```python
import pandas as pd
```

```python
top30_countries = ['United States', 'Spain', 'France', 'United Kingdom', 
                   'Germany', 'Italy', 'Thailand', 'Turkiye', 'Hong Kong SAR China', 
                   'Macao SAR, China', "China (People's Republic of)", 'Canada', 'Japan', 'Austria', 
                   'Netherlands', 'Malaysia', 'Mexico', 'Greece', 'United Arab Emirates', 
                   'Singapore' ,'Russian Federation', 'Poland', 'Saudi Arabia', 
                   'Croatia', 'South Africa', 'Denmark' ,'Czechia', 'Hungary', 
                   'Ukraine', 'Slovak Republic']
```

```python
employement = pd.read_csv('./data/tourism_arrival_employment.csv')
employement
```

```python
employement_topcountries = employement[employement['country'].isin(top30_countries)]
```

```python
employement_topcountries['country'].unique()
```

```python
employement_topcountries.columns
```

```python
columns_to_convert = ['accomodations_service_hotel', 'accomodation_service_other',
       'food_beverage_serving', 'passenger_transport',
       'travel_agency_and_other_resa', 'other_tourism_industry']

def clean_and_convert(column):
    return column.str.replace('\u202f', '').str.replace(',', '').astype(float)

for column in columns_to_convert:
    employement_topcountries[column] = clean_and_convert(employement_topcountries[column])
```

```python
employement_topcountries.info()
```

```python
employement_topcountries[columns_to_convert] = employement_topcountries[columns_to_convert]*1000
```

```python
employement_topcountries.to_csv('./data_visu/03_employment/employment_topcountries.csv', index=False)
```

```python
employement_topcountries = employement_topcountries[employement_topcountries['year']>=2004]
```
