```python
import pandas as pd
```

```python
top30_countries = ['United States', 'Spain', 'France', 'United Kingdom', 
                   'Germany', 'Italy', 'Thailand', 'Turkiye', 'Hong Kong SAR China', 
                   'Macao SAR China', 'China', 'Canada', 'Japan', 'Austria', 
                   'Netherlands', 'Malaysia', 'Mexico', 'Greece', 'United Arab Emirates', 
                   'Singapore' ,'Russian Federation', 'Poland', 'Saudi Arabia', 
                   'Croatia', 'South Africa', 'Denmark' ,'Czechia', 'Hungary', 
                   'Ukraine', 'Slovak Republic']
```

```python
tourism_arrival_departure = pd.read_csv('../data/to use/tourism_arrival_departure.csv')
```

```python
tourism_arrival_departure['year'].unique()
```

```python
df_2019 = tourism_arrival_departure[tourism_arrival_departure['year'] == 2019]
```

```python
df_2019_sorted = df_2019.sort_values(by='arrivals', ascending=False)
```

```python
top_40_arrivals = df_2019_sorted.head(40)
top_40_arrivals = top_40_arrivals.sort_values(by='arrivals', ascending=True)
```

```python
top_40_arrivals['arrival_rank'] = range(1, 41)
top_40_arrivals
```

```python
top_40_arrivals_departure_2019 = top_40_arrivals.sort_values(by='tourist_departures', ascending=True)
```

```python
top_40_arrivals_departure_2019['departure_rank'] = range(1, 41)
top_40_arrivals_departure_2019
```

```python
tourism_arrival_departure.columns
```

```python
columns_to_mean = ['arrivals', 'revenue_dollars', 'gdp_dollars', 'percent_of_gdp', 
                   'primary_energy_consumption_per_capita_kwh/person', 
                   'annual_co2_emissions_per_capita)', 'tourist_departures']

tourism_arrival_departure_mean = tourism_arrival_departure.groupby('country')[columns_to_mean].mean().reset_index()
```

```python
tourism_arrival_departure_mean
```

```python
top30countries_2004_2021 = tourism_arrival_departure_mean.sort_values(by='arrivals', ascending=False)
topa30_arrivals_2004_2021 = top30countries_2004_2021.head(30)
topa30_arrivals_2004_2021 = topa30_arrivals_2004_2021.sort_values(by='arrivals', ascending=True)
topa30_arrivals_2004_2021['arrival_rank'] = range(1, 31)
```

```python
topa30_arrivals_2004_2021
```

```python
topa30_arrivals_2004_2021 = topa30_arrivals_2004_2021.sort_values(by='revenue_dollars', ascending=False)
topa30_arrivals_2004_2021 = topa30_arrivals_2004_2021.head(30)
topa30_arrivals_2004_2021 = topa30_arrivals_2004_2021.sort_values(by='revenue_dollars', ascending=True)
topa30_arrivals_2004_2021['revenue_rank'] = range(1, 31)
```

```python
topa30_arrivals_2004_2021
```

```python
topa30_arrivals_2004_2021.columns
```

```python
topa30_arrivals_2004_2021.to_csv('./data_visu/top30_arrivals_2004_2021.csv', index=False)
```

```python
demography = pd.read_csv('../data/to use/demography.csv')
```

```python
demography = demography[['code','year','population_size_in_millions']]
```

```python
demography['code'].unique()
```

```python
country_code_mapping = {
    'Slovak Republic': 'SVK', 'Ukraine': 'UKR', 'Hungary': 'HUN', 'Czechia': 'CZE', 'Denmark': 'DNK',
    'South Africa': 'ZAF', 'Croatia': 'HRV', 'Saudi Arabia': 'SAU', 'Poland': 'POL',
    'Russian Federation': 'RUS', 'Singapore': 'SGP', 'United Arab Emirates': 'ARE',
    'Greece': 'GRC', 'Mexico': 'MEX', 'Malaysia': 'MYS', 'Netherlands': 'NLD', 'Austria': 'AUT', 'Japan': 'JPN',
    'Canada': 'CAN', 'China': 'CHN', 'Macao SAR, China': 'MAC', 'Hong Kong SAR, China': 'HKG',
    'Turkiye': 'TUR', 'Thailand': 'THA', 'Italy': 'ITA', 'Germany': 'DEU', 'United Kingdom': 'GBR',
    'France': 'FRA', 'Spain': 'ESP', 'United States': 'USA'
}

topa30_arrivals_2004_2021['code'] = topa30_arrivals_2004_2021['country'].map(country_code_mapping)
```

```python
demography_mean = demography.groupby('code').mean().reset_index()
```

```python
demography_mean = demography_mean.drop(columns=['year'])
```

```python
demography_mean
```

```python
topa30_arrivals_2004_2021_2 = pd.merge(topa30_arrivals_2004_2021,demography_mean, on=['code'],how='left')
topa30_arrivals_2004_2021_2.to_csv('./data_visu/top30_arrivals_2004_2021_2.csv', index=False)
```
